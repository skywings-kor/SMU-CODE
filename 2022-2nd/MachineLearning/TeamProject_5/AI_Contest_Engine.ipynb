```python
#ai_contest라는 폴더로 이동합니다
%cd /data/ai_contest

#라이브러리 모듈을 사용합니다(glob이라는거)
from glob import glob

#라벨 데이터들을 이용한 학습을 위하여 학습 테스트 검증으로 3가지로 나눕니다
#train은 학습 이미지
#test는 학습하면서 중간 중간 테스트하여 보완하는 작업을 거칠 때 어느정도 맞추는지 테스트하기 위한 사진
#검증은 학습이 다 끝나고 학습 결과를 보여주기 위한 검증용 이미지
#그래서 비율은 8:1:1로 진행하여 학습을 진행하였음
#8(train):1(test):1(valid)

#test폴더 안에 있는 모든 .jpg 형식의 파일 이름들을 test_img_list에 저장합니다
test_img_list=glob('./test/*.jpg')

#train폴더 안에 있는 모든 .jpg 형식의 파일 이름들을 train_img_list에 저장합니다
train_img_list=glob('./train/*.jpg')

#valid폴더 안에 있는 모든 .jpg 형식의 파일 이름들을 valid_img_list에 저장합니다
valid_img_list=glob('./valid/*.jpg')


#작성하는데 무엇을? test.txt파일에 작성해요.(w=write 구분으로 열어서 해당 파일 이름이 없을경우 생성하고 만일 이미 존재하는 파일인 경우 안에 내용 싹 지우고 처음부터 작성해요)
with open('./test.txt','w') as f:
    #이걸 왜 하냐면 자 학습을 하려면 해당 파일의 경로를 알아야 하는데 test.txt파일에 이제 YOLO가 접근을 하면 test.txt파일에서 상대경로로 작성되어있는 이미지 파일들을 가져와 학습하기 위해서죠
    f.write('\n'.join(test_img_list)+'\n')

#밑에 나머지 2개 것들도 test.txt 파일 작성할 때 같은 이유에서 입니다.
with open('./train.txt','w')as f:
    f.write('\n'.join(train_img_list)+'\n')
    
with open('./valid.txt','w')as f:
    f.write('\n'.join(valid_img_list)+'\n')
```

```python
#자 여기서부터 학습을 진행해요! 
#먼저 yolov5라는 폴더로 이동을 합니다
%cd /data/yolov5

#파이썬 구분을 실행하기 위해 python을 쓰고

#train.py가 무엇이냐면 기존 기계학습 학습 할때 미리 뭐 전처리하고 뭐시기하고 막 귀찮은 것들 과정이 많잖아요 이거 하나에 다 작성되어 있어서 맞춰서 자동으로 되요

#img는 학습 이미지 사이즈는 640 * 640으로 하고

#batch는 학습할 때 한번에 처리할 이미지 수를 말하고

#epochs는 기계학습 수업할 때 알텐데 학습을 몇번 반복할 것인지 작성하는 거고

#data는 그 내가 또 aicontest.yaml라는 것을 만들었는데 거기에 라벨링 기본 정보들이 담겨있음(ex: 라벨1: 사과 , 라벨2:포도, 라벨3:사람) 이런것처럼...
#쨋든 data폴더안에 해당 이름을 가진 파일을 내가 만들어놨고 여기에 있는 라벨 데이터로 참고하는거지

#cfg는 이제 학습할 모델을 정해, (YOLO 5버전은 5가지 모델이 존재해 nano, small, middle, large, xlarge 근데
#nano가 학습이 가장 빠른 대신 convolution layer가 작고 많이 그 수업때 해보면 그 거치는게 적어서 정확도가 매우 떨어져
#xlarge로 갈수록 학습이 느려지고 convolution layer가 더욱 촘촘해져서 정확도가 매우 높아지고 학습률도 매우 좋은 모습을 보이지
#다만 xlarge로 갈수록 더욱 좋은 그래픽카드(GPU)와 좋은 성능의 컴퓨터가 필요함
#나는 여기서 middle 모델을 썼긴했는데 large로 바꿔볼까 고민중..

#weights가 뭐냐면 가중치라는건데 , 너가 만약 학습을 해 그러면 weights 학습이 다된 파일이 해당 뜻을 말하는거야.(best.pt)
#그것을 만일 더 학습을 하고싶어! 그러면 해당 파일에서 이어서 해야하잖아!
#그럴 경우 저기에 해당 이어서 학습시키고 싶은 파일 이름을 넣어서 진행하게 되는데
#나는 지금 처음 학습을 진행하는 거니까 yolov5m.pt라는 기본 그냥 가중치로 처음부터 진행하는거야

#name은 뭐 그냥 다 학습이 끝나면 생기는 폴더 이름을 정하는거인데 이것은 학습 데이터에 필요 없어서 설명은 안할게
!python train.py --img 640 --batch 16 --epochs 500 --data ./data/aicontest.yaml --cfg ./models/yolov5m.yaml --weights yolov5m.pt --name AI_Contest_MiddleModel
```
