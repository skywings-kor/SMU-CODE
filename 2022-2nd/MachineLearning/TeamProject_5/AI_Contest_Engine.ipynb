```python
#ai_contestÎùºÎäî Ìè¥ÎçîÎ°ú Ïù¥ÎèôÌï©ÎãàÎã§
%cd /data/ai_contest

#ÎùºÏù¥Î∏åÎü¨Î¶¨ Î™®ÎìàÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§(globÏù¥ÎùºÎäîÍ±∞)
from glob import glob

#ÎùºÎ≤® Îç∞Ïù¥ÌÑ∞Îì§ÏùÑ Ïù¥Ïö©Ìïú ÌïôÏäµÏùÑ ÏúÑÌïòÏó¨ ÌïôÏäµ ÌÖåÏä§Ìä∏ Í≤ÄÏ¶ùÏúºÎ°ú 3Í∞ÄÏßÄÎ°ú ÎÇòÎàïÎãàÎã§
#trainÏùÄ ÌïôÏäµ Ïù¥ÎØ∏ÏßÄ
#testÎäî ÌïôÏäµÌïòÎ©¥ÏÑú Ï§ëÍ∞Ñ Ï§ëÍ∞Ñ ÌÖåÏä§Ìä∏ÌïòÏó¨ Î≥¥ÏôÑÌïòÎäî ÏûëÏóÖÏùÑ Í±∞Ïπ† Îïå Ïñ¥ÎäêÏ†ïÎèÑ ÎßûÏ∂îÎäîÏßÄ ÌÖåÏä§Ìä∏ÌïòÍ∏∞ ÏúÑÌïú ÏÇ¨ÏßÑ
#Í≤ÄÏ¶ùÏùÄ ÌïôÏäµÏù¥ Îã§ ÎÅùÎÇòÍ≥† ÌïôÏäµ Í≤∞Í≥ºÎ•º Î≥¥Ïó¨Ï£ºÍ∏∞ ÏúÑÌïú Í≤ÄÏ¶ùÏö© Ïù¥ÎØ∏ÏßÄ
#Í∑∏ÎûòÏÑú ÎπÑÏú®ÏùÄ 8:1:1Î°ú ÏßÑÌñâÌïòÏó¨ ÌïôÏäµÏùÑ ÏßÑÌñâÌïòÏòÄÏùå
#8(train):1(test):1(valid)

#testÌè¥Îçî ÏïàÏóê ÏûàÎäî Î™®Îì† .jpg ÌòïÏãùÏùò ÌååÏùº Ïù¥Î¶ÑÎì§ÏùÑ test_img_listÏóê Ï†ÄÏû•Ìï©ÎãàÎã§
test_img_list=glob('./test/*.jpg')

#trainÌè¥Îçî ÏïàÏóê ÏûàÎäî Î™®Îì† .jpg ÌòïÏãùÏùò ÌååÏùº Ïù¥Î¶ÑÎì§ÏùÑ train_img_listÏóê Ï†ÄÏû•Ìï©ÎãàÎã§
train_img_list=glob('./train/*.jpg')

#validÌè¥Îçî ÏïàÏóê ÏûàÎäî Î™®Îì† .jpg ÌòïÏãùÏùò ÌååÏùº Ïù¥Î¶ÑÎì§ÏùÑ valid_img_listÏóê Ï†ÄÏû•Ìï©ÎãàÎã§
valid_img_list=glob('./valid/*.jpg')


#ÏûëÏÑ±ÌïòÎäîÎç∞ Î¨¥ÏóáÏùÑ? test.txtÌååÏùºÏóê ÏûëÏÑ±Ìï¥Ïöî.(w=write Íµ¨Î∂ÑÏúºÎ°ú Ïó¥Ïñ¥ÏÑú Ìï¥Îãπ ÌååÏùº Ïù¥Î¶ÑÏù¥ ÏóÜÏùÑÍ≤ΩÏö∞ ÏÉùÏÑ±ÌïòÍ≥† ÎßåÏùº Ïù¥ÎØ∏ Ï°¥Ïû¨ÌïòÎäî ÌååÏùºÏù∏ Í≤ΩÏö∞ ÏïàÏóê ÎÇ¥Ïö© Ïãπ ÏßÄÏö∞Í≥† Ï≤òÏùåÎ∂ÄÌÑ∞ ÏûëÏÑ±Ìï¥Ïöî)
with open('./test.txt','w') as f:
    #Ïù¥Í±∏ Ïôú ÌïòÎÉêÎ©¥ Ïûê ÌïôÏäµÏùÑ ÌïòÎ†§Î©¥ Ìï¥Îãπ ÌååÏùºÏùò Í≤ΩÎ°úÎ•º ÏïåÏïÑÏïº ÌïòÎäîÎç∞ test.txtÌååÏùºÏóê Ïù¥Ï†ú YOLOÍ∞Ä Ï†ëÍ∑ºÏùÑ ÌïòÎ©¥ test.txtÌååÏùºÏóêÏÑú ÏÉÅÎåÄÍ≤ΩÎ°úÎ°ú ÏûëÏÑ±ÎêòÏñ¥ÏûàÎäî Ïù¥ÎØ∏ÏßÄ ÌååÏùºÎì§ÏùÑ Í∞ÄÏ†∏ÏôÄ ÌïôÏäµÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÏ£†
    f.write('\n'.join(test_img_list)+'\n')

#Î∞ëÏóê ÎÇòÎ®∏ÏßÄ 2Í∞ú Í≤ÉÎì§ÎèÑ test.txt ÌååÏùº ÏûëÏÑ±Ìï† Îïå Í∞ôÏùÄ Ïù¥Ïú†ÏóêÏÑú ÏûÖÎãàÎã§.
with open('./train.txt','w')as f:
    f.write('\n'.join(train_img_list)+'\n')
    
with open('./valid.txt','w')as f:
    f.write('\n'.join(valid_img_list)+'\n')
```

```python
#Ïûê Ïó¨Í∏∞ÏÑúÎ∂ÄÌÑ∞ ÌïôÏäµÏùÑ ÏßÑÌñâÌï¥Ïöî! 
#Î®ºÏ†Ä yolov5ÎùºÎäî Ìè¥ÎçîÎ°ú Ïù¥ÎèôÏùÑ Ìï©ÎãàÎã§
%cd /data/yolov5

#ÌååÏù¥Ïç¨ Íµ¨Î∂ÑÏùÑ Ïã§ÌñâÌïòÍ∏∞ ÏúÑÌï¥ pythonÏùÑ Ïì∞Í≥†

#train.pyÍ∞Ä Î¨¥ÏóáÏù¥ÎÉêÎ©¥ Í∏∞Ï°¥ Í∏∞Í≥ÑÌïôÏäµ ÌïôÏäµ Ìï†Îïå ÎØ∏Î¶¨ Î≠ê Ï†ÑÏ≤òÎ¶¨ÌïòÍ≥† Î≠êÏãúÍ∏∞ÌïòÍ≥† Îßâ Í∑ÄÏ∞ÆÏùÄ Í≤ÉÎì§ Í≥ºÏ†ïÏù¥ ÎßéÏûñÏïÑÏöî Ïù¥Í±∞ ÌïòÎÇòÏóê Îã§ ÏûëÏÑ±ÎêòÏñ¥ ÏûàÏñ¥ÏÑú ÎßûÏ∂∞ÏÑú ÏûêÎèôÏúºÎ°ú ÎêòÏöî

#imgÎäî ÌïôÏäµ Ïù¥ÎØ∏ÏßÄ ÏÇ¨Ïù¥Ï¶àÎäî 640 * 640ÏúºÎ°ú ÌïòÍ≥†

#batchÎäî ÌïôÏäµÌï† Îïå ÌïúÎ≤àÏóê Ï≤òÎ¶¨Ìï† Ïù¥ÎØ∏ÏßÄ ÏàòÎ•º ÎßêÌïòÍ≥†

#epochsÎäî Í∏∞Í≥ÑÌïôÏäµ ÏàòÏóÖÌï† Îïå ÏïåÌÖêÎç∞ ÌïôÏäµÏùÑ Î™áÎ≤à Î∞òÎ≥µÌï† Í≤ÉÏù∏ÏßÄ ÏûëÏÑ±ÌïòÎäî Í±∞Í≥†

#dataÎäî Í∑∏ ÎÇ¥Í∞Ä Îòê aicontest.yamlÎùºÎäî Í≤ÉÏùÑ ÎßåÎì§ÏóàÎäîÎç∞ Í±∞Í∏∞Ïóê ÎùºÎ≤®ÎßÅ Í∏∞Î≥∏ Ï†ïÎ≥¥Îì§Ïù¥ Îã¥Í≤®ÏûàÏùå(ex: ÎùºÎ≤®1: ÏÇ¨Í≥º , ÎùºÎ≤®2:Ìè¨ÎèÑ, ÎùºÎ≤®3:ÏÇ¨Îûå) Ïù¥Îü∞Í≤ÉÏ≤òÎüº...
#Ï®ãÎì† dataÌè¥ÎçîÏïàÏóê Ìï¥Îãπ Ïù¥Î¶ÑÏùÑ Í∞ÄÏßÑ ÌååÏùºÏùÑ ÎÇ¥Í∞Ä ÎßåÎì§Ïñ¥ÎÜ®Í≥† Ïó¨Í∏∞Ïóê ÏûàÎäî ÎùºÎ≤® Îç∞Ïù¥ÌÑ∞Î°ú Ï∞∏Í≥†ÌïòÎäîÍ±∞ÏßÄ

#cfgÎäî Ïù¥Ï†ú ÌïôÏäµÌï† Î™®Îç∏ÏùÑ Ï†ïÌï¥, (YOLO 5Î≤ÑÏ†ÑÏùÄ 5Í∞ÄÏßÄ Î™®Îç∏Ïù¥ Ï°¥Ïû¨Ìï¥ nano, small, middle, large, xlarge Í∑ºÎç∞
#nanoÍ∞Ä ÌïôÏäµÏù¥ Í∞ÄÏû• Îπ†Î•∏ ÎåÄÏã† convolution layerÍ∞Ä ÏûëÍ≥† ÎßéÏù¥ Í∑∏ ÏàòÏóÖÎïå Ìï¥Î≥¥Î©¥ Í∑∏ Í±∞ÏπòÎäîÍ≤å Ï†ÅÏñ¥ÏÑú Ï†ïÌôïÎèÑÍ∞Ä Îß§Ïö∞ Îñ®Ïñ¥Ï†∏
#xlargeÎ°ú Í∞àÏàòÎ°ù ÌïôÏäµÏù¥ ÎäêÎ†§ÏßÄÍ≥† convolution layerÍ∞Ä ÎçîÏö± Ï¥òÏ¥òÌï¥Ï†∏ÏÑú Ï†ïÌôïÎèÑÍ∞Ä Îß§Ïö∞ ÎÜíÏïÑÏßÄÍ≥† ÌïôÏäµÎ•†ÎèÑ Îß§Ïö∞ Ï¢ãÏùÄ Î™®ÏäµÏùÑ Î≥¥Ïù¥ÏßÄ
#Îã§Îßå xlargeÎ°ú Í∞àÏàòÎ°ù ÎçîÏö± Ï¢ãÏùÄ Í∑∏ÎûòÌîΩÏπ¥Îìú(GPU)ÏôÄ Ï¢ãÏùÄ ÏÑ±Îä•Ïùò Ïª¥Ìì®ÌÑ∞Í∞Ä ÌïÑÏöîÌï®
#ÎÇòÎäî Ïó¨Í∏∞ÏÑú middle Î™®Îç∏ÏùÑ ÏçºÍ∏¥ÌñàÎäîÎç∞ largeÎ°ú Î∞îÍøîÎ≥ºÍπå Í≥†ÎØºÏ§ë..

#weightsÍ∞Ä Î≠êÎÉêÎ©¥ Í∞ÄÏ§ëÏπòÎùºÎäîÍ±¥Îç∞ , ÎÑàÍ∞Ä ÎßåÏïΩ ÌïôÏäµÏùÑ Ìï¥ Í∑∏Îü¨Î©¥ weights ÌïôÏäµÏù¥ Îã§Îêú ÌååÏùºÏù¥ Ìï¥Îãπ ÎúªÏùÑ ÎßêÌïòÎäîÍ±∞Ïïº.(best.pt)
#Í∑∏Í≤ÉÏùÑ ÎßåÏùº Îçî ÌïôÏäµÏùÑ ÌïòÍ≥†Ïã∂Ïñ¥! Í∑∏Îü¨Î©¥ Ìï¥Îãπ ÌååÏùºÏóêÏÑú Ïù¥Ïñ¥ÏÑú Ìï¥ÏïºÌïòÏûñÏïÑ!
#Í∑∏Îü¥ Í≤ΩÏö∞ Ï†ÄÍ∏∞Ïóê Ìï¥Îãπ Ïù¥Ïñ¥ÏÑú ÌïôÏäµÏãúÌÇ§Í≥† Ïã∂ÏùÄ ÌååÏùº Ïù¥Î¶ÑÏùÑ ÎÑ£Ïñ¥ÏÑú ÏßÑÌñâÌïòÍ≤å ÎêòÎäîÎç∞
#ÎÇòÎäî ÏßÄÍ∏à Ï≤òÏùå ÌïôÏäµÏùÑ ÏßÑÌñâÌïòÎäî Í±∞ÎãàÍπå yolov5m.ptÎùºÎäî Í∏∞Î≥∏ Í∑∏ÎÉ• Í∞ÄÏ§ëÏπòÎ°ú Ï≤òÏùåÎ∂ÄÌÑ∞ ÏßÑÌñâÌïòÎäîÍ±∞Ïïº

#nameÏùÄ Î≠ê Í∑∏ÎÉ• Îã§ ÌïôÏäµÏù¥ ÎÅùÎÇòÎ©¥ ÏÉùÍ∏∞Îäî Ìè¥Îçî Ïù¥Î¶ÑÏùÑ Ï†ïÌïòÎäîÍ±∞Ïù∏Îç∞ Ïù¥Í≤ÉÏùÄ ÌïôÏäµ Îç∞Ïù¥ÌÑ∞Ïóê ÌïÑÏöî ÏóÜÏñ¥ÏÑú ÏÑ§Î™ÖÏùÄ ÏïàÌï†Í≤å
!python train.py --img 640 --batch 16 --epochs 500 --data ./data/aicontest.yaml --cfg ./models/yolov5m.yaml --weights yolov5m.pt --name AI_Contest_MiddleModel
```
