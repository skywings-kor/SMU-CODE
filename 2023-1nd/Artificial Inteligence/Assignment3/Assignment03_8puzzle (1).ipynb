```python
POPULATION_SIZE=10  #염색체 개수
SIZE=8 #이게 염색체 원소 개수인데 , 8-Queen에서의 Queen의 개수임
MUTATION_RATE= 0.07 #여기가 돌연변이 확률인데 그러면 마주칠 확률로해야하나..? 퀸끼리... 일단 0.07로 임의 Set구성
import random
```

```python
class Chromosome:
  #__init__이라는 기능을 만들었습니다
  def __init__(self,g=[]): 
    self.genes=g.copy()    #유전자 리스트로 구현된다는 뜻(?)
    self.fitness=0  #적합도  <= 이거 써서 구현하기 밑에!! 규민아 기억해!!

    if self.genes.__len__()==0: #염색체가 초기 상태이면 초기화
      i=0
      while i<SIZE:
        if random.random() >= 0.5:
          self.genes.append(1)
        
        else:
          self.genes.append(0)
        i+=1
  
  #서로 공격하지 않을 때 -> 28-h..
  #일단 먼저 교재에 나와있는 28-h로 해보겠습니다.
  #교재안처럼 28-h 로 할경우...
  # def cal_fitness(self):  #적합도 계산
  #   return 28 - sum(abs(self.genes[i] - self.genes[j]) == j - i for i in range(8) for j in range(i + 1, 8))

  #제가 한 것은 서로 공격이 가능한 퀸들의 쌍의 개수를 이용한 적합도 기능을 구현한 것입니다.
  #위 교재에서 한 것은 서로를 공격하지 않는거고 이건 일단 서로 공격이 가능한 쌍의 수 입니다.
  #원래는 행, 열 구분하여 하나씩 function 만든 후 후에 적합도 1 0 구분하는걸 할까 하다가 일단 이렇게 해보았습니다.
  def cal_fitness(self):
    self.fitness = 0
    #일단 유전자 배열에 만들어진 만큼 돌게하고...(행)
    for i in range(len(self.genes)):
        #일단 유전자 배열에 만들어진 만큼 돌게 하고...(열)
        for j in range(i+1, len(self.genes)):
            #만일 퀸이 서로 같은 행이거나 서로 같은 대각선 상에 위치해있으면...
            if self.genes[i] == self.genes[j] or abs(i-j) == abs(self.genes[i]-self.genes[j]):
                #적합도 +1
                self.fitness += 1
    #적합도 계산한거 돌려주기          
    return self.fitness

  def __str__(self):
    return self.genes.__str__()
```

```python
#염색체와 적합도 출력
def print_p(pop):
  i=0
  for x in pop:
    print("염색체 #",i,"=",x ,"적합도=",x.cal_fitness() )#위에서 만든 cal_fitness 이름의 function run
    i+=1
  print("")

#선택 연산
def select(pop):
  max_value=sum([c.cal_fitness() for c in population])
  pick = random.uniform(0,max_value)
  current=0

  for c in pop:
    current += c.cal_fitness()
    if current>pick:
      return c
```

```python
#교차 연산
def crossover(pop):
  father=select(pop)
  mother=select(pop)
  index=random.randint(1,SIZE-2)
  child1=father.genes[:index]+mother.genes[index:]
  child2=mother.genes[:index]+father.genes[index:]
  return(child1,child2)

#돌연변이 연산
def mutate(c):
  for i in range(SIZE):
    if random.random()<MUTATION_RATE:
      if random.random()<0.5:
        c.genes[i]=1
      else:
        c.genes[i]=0
```

```python
population=[]
i=0

#초기 염색체를 생성하여 객체 집단에 추가
while i<POPULATION_SIZE:
  population.append(Chromosome()) #Chromosome위에 만들었던거 실행시켜서 population 빈 배열에 추가
  i+=1


count=0
population.sort(key=lambda x:x.cal_fitness(),reverse=True)
print("세대번호=> ",count)
print_p(population)
count=1
```

```python
#자 여기서 기존 써있던 31은  적합도 해당 숫자에 도착하면 되도록 하면 되겠구나
# 그러면 이것을 변수로 바꾸면..? 일단 goalNum이라는 변수를 만들겠습니다.

#각 퀸이 놓인 대각선 상에 퀸이 있는 경우의 최대 수 => 28
goalNum=28

while population[0].cal_fitness()<goalNum:
  new_pop=[]

  #선택과 교차연산
  for _ in range(POPULATION_SIZE//2):
    c1,c2 = crossover(population)
    new_pop.append(Chromosome(c1))
    new_pop.append(Chromosome(c2))

  #자식 세대가 부모세대 대체
  #깊은 복사 수행
  population = new_pop.copy()

  #돌연변이 연산
  for c in population:
    mutate(c)

  #출력을 위한 정렬
  population.sort(key=lambda x:x.cal_fitness(),reverse=True)
  print('세대번호=',count)
  print_p(population)
  count+=1
  if count>100:
    break;
```
